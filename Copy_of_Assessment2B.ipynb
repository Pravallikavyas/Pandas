###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
toppername=[]
a=clg_m[clg_m["Department"]=="Computer Science and Engineering"]
mcs=a["CodeKata Score"].max()
toppername.append(a.Name[a["CodeKata Score"]==a["CodeKata Score"].max()].values[0])
b=clg_m[clg_m["Department"]=="Electronics and Communication Engineering"]
me=b["CodeKata Score"].max()
toppername.append(b.Name[b["CodeKata Score"]==b["CodeKata Score"].max()].values[0])
c=clg_m[clg_m["Department"]=="Electronics and Electrical Engineering"]
mee=c["CodeKata Score"].max()
toppername.append(c.Name[c["CodeKata Score"]==c["CodeKata Score"].max()].values[0])
print(toppername)
```
